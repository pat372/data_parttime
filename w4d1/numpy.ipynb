```python
a = np.random.randn(2000, 3000)
b = np.empty((2000, 3000))
def test_jm(x, y):
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if x[i, j] > 0:
                y[i, j] = 1
            else:
                y[i, j] = -1
```

```python
%timeit test_jm(a, b)
```

```python
a = np.random.randn(2000, 3000)
b = np.empty((2000, 3000))
def test_jm_2(x, y):
    y[x > 0] = 1
    y[x < 0] = -1
```

```python
%timeit test_jm_2(a, b)
```
