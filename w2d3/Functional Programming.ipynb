```python
# functions as simple tools
# idea : stack functions to make a data pipeline
# beware of having state being changed in a function, we want a function to be "pure"
```

```python
# pure function :
def count_names(l):
    return len(l)

# not a pure function
count = 0
def increment_count(l):
    global count
    count += len(l)
    return count

a = ['alice', 'bob', 'charles']
print(count_names(a))
print(increment_count(a))
print(increment_count(a))
```
